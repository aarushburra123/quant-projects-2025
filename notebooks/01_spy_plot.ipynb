```python
import yfinance as yf, matplotlib.pyplot as plt
spy = yf.download("SPY", start="2000-01-01", auto_adjust=True)
spy['Close'].plot(title="SPY Close 2000–present")
plt.show()
```

```python
# This will display monthly returns from the adjusted close
monthly = spy['Close'].resample('M').last().pct_change().dropna()
sharpe = (monthly.mean() / monthly.std()) * (12 ** 0.5)  # simple annualized Sharpe, Rf≈0
print("Monthly mean:", round(monthly.mean(), 4), "Monthly std:", round(monthly.std(), 4))
print("Annualized Sharpe:", round(sharpe, 2))
monthly.plot(title="SPY Monthly Returns"); plt.show()
```

```python
plt.tight_layout()
plt.savefig("results/spy_plot.png", dpi=150)
#png ofthe spy plot will go here
```
